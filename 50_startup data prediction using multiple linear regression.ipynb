## Encode the categorical variables using one-hot encoding

```python
newdf = pd.get_dummies(df,dtype=int)
newdf
```

## Perform feature scaling on numerical features.

```python
from sklearn.preprocessing import StandardScaler
x=['R&D Spend','Administration','Marketing Spend']
scaler = StandardScaler()
scaled_x=scaler.fit_transform(newdf[x])
```

## Assign X and Y variables respectively

```python
X=pd.DataFrame(scaled_x,columns=x)
X=X.join(newdf[['State_California','State_Florida','State_New York']])
Y=newdf['Profit']
X.head()
```

```python
Y.head()
```

## Split the data into training and testing

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,Y, test_size = 0.3, random_state=100)
```

## MODEL BUILDING

## Create a multiple linear regression model

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X_train, y_train)
```

## Check training score of model

```python
model.score(X_train, y_train)
```

## Predict the model

```python
y_pred = model.predict(X_test)
```

## Evaluate the testing data using mean squared error

```python
from sklearn import metrics
print ('Mean Squared Error is: ', metrics.mean_squared_error(y_test, y_pred))
```
